```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
```

```python
train_tests = []
for train_index, test_index in skf.split(df["Embedded_text"], df["majority_sentiment"]):
  coll = [train_index, test_index]
  train_tests.append(coll)
```

```python
i = 0
train_tests_df = []
while i < 5:
  currentSelection = train_tests[i]
  train_df = df.iloc[currentSelection[0]]
  test_df = df.iloc[currentSelection[1]]

  train_tests_df.append([train_df, test_df])
  
  i = i + 1
```
