```python
import statsmodels
from statsmodels.stats import inter_rater as irr
import krippendorff
```

```python
df = anno_all[["Sentiment", "Sentiment2", "Sentiment3"]]
df = df.to_numpy()
df
```

```python
agg = irr.aggregate_raters(df) # returns a tuple (data, categories)
agg
```

```python
irr.fleiss_kappa(agg[0], method='fleiss')
```
